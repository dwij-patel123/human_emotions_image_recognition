```python
import numpy as np
import pandas as pd
import tensorflow as tf
import os
import cv2
import sklearn
import matplotlib.pyplot as plt
```

```python
def input_train_test(DATADIR,CATEGORIES):
    data = []
    
    for category in CATEGORIES:
        path = os.path.join(DATADIR,category)
        label = CATEGORIES.index(category)
        for img in os.listdir(path):
            print(img)
            img_arr = cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
            new_arr = cv2.resize(img_arr,(224,224))
            data.append([new_arr,label])
            
    return data


DATADIR_TRAIN = "C:\dwij\deep learning projects\human_emotions_dataset\Emotions_Dataset\Train_data"

DATADIR_TEST = "C:\dwij\deep learning projects\human_emotions_dataset\Emotions_Dataset\Test_data"

CATEGORIES = ["angry","happy","sad"]

training_data = input_train_test(DATADIR_TRAIN,CATEGORIES)

testing_data = input_train_test(DATADIR_TEST,CATEGORIES)

```

```python
# distributiong data into X_train and y_train

def train_test_data(input_data):
    X = []
    y = []
    
    for data in input_data:
        X.append(data[0])
        y.append(data[1])
    
    return X,y
```

```python
X_train,y_train = train_test_data(training_data)
X_test,y_test = train_test_data(testing_data)
```

```python
X_train = np.array(X_train)
y_train = np.array(y_train)
X_train = np.expand_dims(X_train,axis=-1)
X_test = np.array(X_test)
y_test = np.array(y_test)
X_test = np.expand_dims(X_test,axis=-1)
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32,kernel_size=5,strides=1,activation='relu',input_shape=(224,224,1)),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPool2D(2),
    tf.keras.layers.Dropout(0.3),
    
    
    tf.keras.layers.Conv2D(16,kernel_size=5,strides=1,activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPool2D(2),
    tf.keras.layers.Dropout(0.2),
    
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(86,activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(3,activation='softmax')
])

model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X,y,epochs=6)
```

```python
model.evaluate(X_test,y_test)
```
